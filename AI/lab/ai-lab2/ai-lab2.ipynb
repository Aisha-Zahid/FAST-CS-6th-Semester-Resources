# Write a Program for a Library System:
Ask user about his name, create a dictionary to store his name and roll number, if his
roll number exist, say him Hi with his name and give him choices to select from
different books category after selecting books category and then books at the end
give him all the information which book he selects and random time to return books.
▪ Use classes for writing program.

```python
class Library:
    def __init__(self):
        self.users = {}  
        self.books = {
            "Tech": ["C++", "Python", "JavaScript"],
            "Educational": ["Algorithms", "English", "Science"],
        }

    def add_user(self, roll, name):
        self.users[roll] = name

    def greet_user(self, roll):
        if roll in self.users:
            print(f"Hey! {self.users[roll]}!")
            return True
        else:
            print("Sorry rollno. not found")
            return False

    def select_category(self):
        print("Choose book category:")
        for idx, category in enumerate(self.books.keys()):
            print(f"{idx + 1}. {category}")
        choice = int(input("Enter category : "))
        categories = list(self.books.keys())
        if 1 <= choice <= len(categories):
            return categories[choice - 1]
        else:
            print("Invalid choice")
            

    def select_book(self, category):
        print(f"Choose a book from {category}:")
        for idx, book in enumerate(self.books[category]):
            print(f"{idx + 1}. {book}")
        choice = int(input("Enter book : "))
        if 1 <= choice <= len(self.books[category]):
            return self.books[category][choice - 1]
        else:
            print("Invalid choice")
            

    def generate_return_date(self):
        return datetime.now() + timedelta(days=random.randint(1,1))

def main():
    lib = Library()

    # users in listed
    lib.add_user("9136", "Ayesha")
    lib.add_user("9777", "Samia")
    lib.add_user("6923", "Fatima")

    # user input for roll number
    roll = input("Enter your roll number: ")

    # condition if roll no exsist
    if lib.greet_user(roll):
        # Select book category
        category = lib.select_category()
        if category:
            # books selection 
            book = lib.select_book(category)
            if book:
                print(f"You selected {book} ")
                return_date = lib.generate_return_date()
                print(f"Please return the book by {return_date.strftime('%Y-%m-%d')} ")

if __name__ == "__main__":
    main()
```

# Implement the following graph with dictionary and List.

```python
graph = {
    1: [2, 5],
    2: [1, 3, 5],
    3: [2, 4],
    4: [5, 6],
    5: [1, 2, 4],
    6: [4]
}


graph_list = [
    [],          # Node 0 
    [2, 5],      # 1
    [1, 3, 5],   # 2
    [2, 4],      # 3
    [5, 6],      # 4
    [1, 2, 4],   # 5
    [4]          # 6
]

# Printing the graph
print("Graph represented using dictionary:")
for node, neighbors in graph.items():
    print(f"Node {node} is connected to: {neighbors}")

print("Graph represented using lists:")
for node, neighbors in enumerate(graph_list):
    print(f"Node {node} is connected to: {neighbors}")
```
